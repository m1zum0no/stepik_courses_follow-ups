```python
df = pd.read_csv(('emg.dat'), 
sep="\s+", 
skiprows=1, 
usecols=[0,1], 
names=['TIME','VALUE'])
time = df[df.columns[0]]
emg = df[df.columns[1]]
emg.head(n = 5)
```

```python
fig = plt.figure()
plt.plot(time, emg)
plt.xlabel('Time (sec)')
plt.ylabel('EMG (a.u.)')
fig_name = 'fig2.png'
fig.set_size_inches(w=11,h=7)
fig.savefig(fig_name)
```
